 ## Key Performance Indicators (KPIs)

```python
# Average Anxiety, Depression, Stress, Loneliness
avg_anxiety = df['Anxiety Score'].mean()
avg_depression = df['Depression Score'].mean()
avg_stress = df['Stress Level'].mean()
avg_loneliness = df['Loneliness Score'].mean()
```

```python
print(f"Average Anxiety Score: {avg_anxiety:.2f}")
print(f"Average Depression Score: {avg_depression:.2f}")
print(f"Average Stress Level: {avg_stress:.2f}")
print(f"Average Loneliness Score: {avg_loneliness:.2f}")
```

```python
# Percentage with High Anxiety/Depression/Stress
high_anxiety_threshold = df['Anxiety Score'].quantile(0.75)
high_depression_threshold = df['Depression Score'].quantile(0.75)
high_stress_threshold = df['Stress Level'].quantile(0.75)
high_loneliness_threshold = df['Loneliness Score'].quantile(0.75)
```

```python
high_anxiety_percent = (df['Anxiety Score'] > high_anxiety_threshold).mean() * 100
high_depression_percent = (df['Depression Score'] > high_depression_threshold).mean() * 100
high_stress_percent = (df['Stress Level'] > high_stress_threshold).mean() * 100
high_loneliness_percent = (df['Loneliness Score'] > high_loneliness_threshold).mean() * 100

print(f"Percentage with High Anxiety: {high_anxiety_percent:.2f}%")
print(f"Percentage with High Depression: {high_depression_percent:.2f}%")
print(f"Percentage with High Stress: {high_stress_percent:.2f}%")
print(f"Percentage with High Loneliness: {high_loneliness_percent:.2f}%")
```

```python
# Impact of Sleep and Physical Activity on Mental Health
print(f"Correlation Sleep vs Anxiety: {df['Sleep Hours'].corr(df['Anxiety Score']):.2f}")
print(f"Correlation Physical Activity vs Depression: {df['Physical Activity Hrs'].corr(df['Depression Score']):2f}")

# Impact of social support on mental health.
print(f"Correlation Social Support vs Anxiety: {df['Social Support Score'].corr(df['Anxiety Score']):.2f}")
print(f"Correlation Social Support vs Depression: {df['Social Support Score'].corr(df['Depression Score']):.2f}")
```

```python
# Impact of therapy and medication
print(f"Average Anxiety of people in Therapy: {df[df['Therapy']=='Yes']['Anxiety Score'].mean():.2f}")
print(f"Average Anxiety of people using medication: {df[df['Medication Use']=='Yes']['Anxiety Score'].mean():.2f}")
```
